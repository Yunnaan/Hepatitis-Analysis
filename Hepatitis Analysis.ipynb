```python
df['class'] = np.where(df['class'] == 'live', 0, 1)
df.head()                                             # Encoding class column
```

```python
df['sex'] = np.where(df['sex'] == 'live', 0, 1)
df.head()                                        # Encoding sex column
```

```python
df ['class'].value_counts()
```

```python
x = df.drop (['class'], axis = 1)
y = df ['class']                     # spliting the dataset into x and y 
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
# importing train and test data set for training
```

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.2 , stratify = y, random_state = 2)
# splitting dataset into train and test subset
```

# MODEL TRAINING

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression ()
model.fit (x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
x_train_prediction = model.predict (x_train)
training_data_accuracy = accuracy_score (x_train_prediction, y_train)
print ("Accuracy Score On Training Data:", training_data_accuracy)
```

```python
train_score = model.score (x_train, y_train)
print ('Accuracy Score on Training Data:', train_score)
```

```python
test_score = model.score (x_test, y_test)
print ("Accuracy Score On Test Data:", test_score)
```

# BUILDING A PREDICTION SYSTEM

```python
input_data = (42,False,False,False,False,False,True,False,False,False,False,False,1,85,14,4,100,False,2
)

data = np.array(input_data)

input_data_reshaped = data.reshape (1, -1)

prediction = model.predict (input_data_reshaped)

print (prediction)

if prediction == 0:
    print ('This Person may still live despired having hepatitis')
else:
    print ('This Person might likely die of hepatitis')
```
